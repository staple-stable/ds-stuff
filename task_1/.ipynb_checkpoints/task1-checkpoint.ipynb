Корреляция признаков:

```python
data.corr()
```

```python
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(data.corr(), vmax=.8, square=True, annot=True);
```

Из интересных сильных корреляций:
- SAT обратно коррелирует с процентом поступивших. Довольно логично, чем выше балл, тем сложнее поступить.
- Также, SAT прямо кореллирует со стоимостью обучения, процентом выпускников(немного неожиданно), top10% студентами.
- Acceptance обратно коррелирует со всеми остальными признаками. Чем выше требования - меньше поступающих. PhD и стоимость затронуты этим гораздо меньше.
- К слову о PhD, максимальную корреляцию он показал с Top10%. Немного неочевидно, ожидал большей корреляции с силой университета(т.е. уровень студентов и стоимость). Впрочем они идут следом.
- С ростом числа поступивших, становится меньше выпустившихся.
